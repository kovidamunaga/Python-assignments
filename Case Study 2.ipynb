```python
import os
import pandas as pd
import numpy as np
```

```python
case2=pd.read_csv('Store.csv',encoding="latin")
```

```python
case2
```

```python
## How many unique cities are the orders being delivered to
case2.City.unique()
```

```python
## What is the total quantity sold in the East Region?
case2[(case2.Region=="East")][['Quantity']].count()
```

```python
ff=case2[(case2.Region=="East")][['Quantity']]
ff.shape[0]
```

```python
## Find the sum of the quantity sold in the East Region
case2[case2.Region=="East"][['Quantity']].sum()
```

```python
## In the south region sort the sales in decreasing order
case2[case2.Region=="South"][['Sales','Region']].sort_values(by='Sales')
```

```python
## Find the mean of quantity for every region
case2.groupby('Region').agg({'Quantity':np.mean})
```

```python
## Find the mean of sales for every category
case2.groupby("Category").agg({'Sales':np.mean})


```

```python
## Find the max, min, sum of sales and profit for every category
case2.groupby("Category")[['Sales','Profit']].agg(['max','min','sum'])
```

```python
## Find sum of sales and max profit for every segment
case2.groupby('Segment')[['Sales','Profit']].agg({'Profit':np.max,'Sales':np.sum})
```

```python
## For every segment find the mean of the discount
case2.groupby('Segment')[['Discount']].mean()
```

```python
## For every segment find the most profitable customers
case2[["Customer Name","Profit"]].loc[case2.groupby(['Segment'])["Profit"].idxmax()] 
```

```python
## What are the top 5 categories that give maximum profit?

case2.groupby('Sub-Category')[['Profit']].max().nlargest(5, 'Profit')




```

```python
## What is the Total Sales, Quantity, Discount, Profit across Total US.
case2[case2.Country=='United States'][['Sales','Quantity','Discount','Profit']].sum()
```

```python
## How many times has it taken more than 5 days from placing an order to shipping

case2['Order Date']=pd.to_datetime(case2['Order Date'])
case2['Ship Date']=pd.to_datetime(case2['Ship Date'])
```

```python
case2['days']=case2['Ship Date']-case2['Order Date']
sum(case2.days > '5days')
```

```python
case2[case2.days>'5 days'][['days']].count()

case2
```

```python
## Find the total number of orders in every category which has been shipped with a duration > 5 days

case2[case2.days>'5 days'].groupby('Category')[['Category']].agg('count')


```

```python
## What’s the percentage of items which has been shipped within 5 days

case2[case2.days<'5 days'].shape[0]/case2.shape[0]*100

```

```python
## What’s the percentage of items which has been shipped after 5 days
case2[case2.days>'5 days'].shape[0]/case2.shape[0]*100
```
