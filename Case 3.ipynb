```python
## Q.7 ) label all the incidents where the number killed was more than 5 as severe.  
elem=[]
for i in range(0,len(case3)):
    if case3.iloc[i]['nkill'] > 5:
        elem.append('severe')
    else:
        elem.append('not severe')
case3['incidents']=elem
print(case3['incidents'].head())
        
```

```python
## Q.9) Create a new category representing if the incident occured in Afghanistan, Pakistan or India as one
#level of the category and all the other countries as another level

case3['inc_category']=['category1'if (case3.iloc[i]['country_txt']=='Pakistan'
                                       or case3.iloc[i]['country_txt']=='India'
                                       or case3.iloc[i]['country_txt']=='Afghanistan') else 'category2' for i in range(0,len(case3))]

print(case3['inc_category'].head())
```
